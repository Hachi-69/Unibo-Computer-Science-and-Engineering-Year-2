## Esercizio 0
- Costruire due array numpy-array 2d di dimensione 3000x3000 i cui valori sono campionati da una distribuzione gaussiana con media 0 e varianza 1.
- Effetuare la somma tra le due matrici con due metodi:

        -  utilizzando le operazioni vettoriali 
        -  considerando due cicli espliciti annidati, su righe e colonne
- Confrontare le prestazioni in termini di tempo di esecuzione (utilizzando perf_counter() del modulo time)
        


```python
import numpy as np
import time

N = 3000

A = np.random.normal(loc=0.0, scale=1.0, size=(N, N))
B = np.random.normal(loc=0.0, scale=1.0, size=(N, N))

# Operazioni Vettoriali, Metodo Corretto

start_vec = time.perf_counter()

somma = A + B

end_vec = time.perf_counter()
tempo_vec = end_vec - start_vec
print(f"Tempo di esecuzione (Vettoriale): {tempo_vec:.5f} secondi")

# Cicli Espliciti Annidati

C_loop = np.zeros((N, N))

start_loop = time.perf_counter()

for i in range(N):
    for j in range(N):
        C_loop[i, j] = A[i, j] + B[i, j]

end_loop = time.perf_counter()
tempo_loop = end_loop - start_loop
print(f"Tempo di esecuzione (Cicli annidati): {tempo_loop:.5f} secondi")

# Calcolo e stampo il guadagno prestazionale (Speedup)
speedup = tempo_loop / tempo_vec
print(f"\nLa vettorializzazione è stata {speedup:.2f} volte più veloce!")
```

- Verificare, usando l'attributo `.base`, se l'array risultante dal `reshape` è una vista o una copia dell'array originale.

```python
import numpy as np

s = np.linspace(0, 20, 40)
s_reshaped = s.reshape(4, -1)

if np.shares_memory(s, s_reshaped):
    print("s_reshaped è una VISTA di s")
else:
    print("s_reshaped è una COPIA di s")
```

 ## Esercizio 8
-     Scrivere il numpy  array 2d, A, [[10,11,12],[13,14,15],[16,17,18]] 
-     trasformarlo in un numpy array 1d usando il metodo .ravel()  
-     Si tratta di una copia o di una vista?

```python
import numpy as np

array = np.array([[10, 11, 12],
                  [13, 14, 15],
                  [16, 17, 18]])

print("Array originale A (2D):")
print(array)
print(f"Shape: {array.shape}\n")
print("-" * 30)

array_1d = array.ravel()

print("Array appiattito A_1d (1D):")
print(array_1d)
print(f"Shape: {array_1d.shape}\n")
print("-" * 30)

if np.shares_memory(array, array_1d):
    print("Si tratta di una view.")
else:
    print("Si tratta di una copy.")
```
