## Esercizio 0
- Costruire due array numpy-array 2d di dimensione 3000x3000 i cui valori sono campionati da una distribuzione gaussiana con media 0 e varianza 1.
- Effetuare la somma tra le due matrici con due metodi:

        -  utilizzando le operazioni vettoriali 
        -  considerando due cicli espliciti annidati, su righe e colonne
- Confrontare le prestazioni in termini di tempo di esecuzione (utilizzando perf_counter() del modulo time)
        


```python
import numpy as np
import time

N = 3000

A = np.random.normal(loc=0.0, scale=1.0, size=(N, N))
B = np.random.normal(loc=0.0, scale=1.0, size=(N, N))

# Operazioni Vettoriali, Metodo Corretto

start_vec = time.perf_counter()

somma = A + B

end_vec = time.perf_counter()
tempo_vec = end_vec - start_vec
print(f"Tempo di esecuzione (Vettoriale): {tempo_vec:.5f} secondi")

# Cicli Espliciti Annidati

C_loop = np.zeros((N, N))

start_loop = time.perf_counter()

for i in range(N):
    for j in range(N):
        C_loop[i, j] = A[i, j] + B[i, j]

end_loop = time.perf_counter()
tempo_loop = end_loop - start_loop
print(f"Tempo di esecuzione (Cicli annidati): {tempo_loop:.5f} secondi")

# Calcolo e stampo il guadagno prestazionale (Speedup)
speedup = tempo_loop / tempo_vec
print(f"\nLa vettorializzazione è stata {speedup:.2f} volte più veloce!")
```

## Esercizio 26
Scrivi un programma Python che esegua l'elaborazione di un'immagine utilizzando un filtro di media. 

Utilizza il padding per gestire i pixel del bordo durante l'applicazione del filtro.

Il programma deve seguire questi passaggi:

- Carica l'immagine: Carica l'immagine 'cameraman.png' utilizzando la libreria Matplotlib.**image=plt.imread('cameraman.png')**
- Visualizza l'immagine originale **plt.imshow(image,cmap='gray')**
- definisce la dimensione del filtro, dim_filtro
- Applica il padding: Applica un padding costante di dim_filtro//2 pixel a tutti i bordi dell'immagine (utilizzando il metodo di numpy **image_bordered=np.pad(image,(dim_filtro//2,dim_filtro//2),, mode='reflect')**
- Crea il filtro di media: Crea un filtro di media di dimensioni *dim_filtroxdim_filtro*, dove tutti gli elementi sono pari a 1/(dim_filtroxdim_filtro)
- Applica il filtro: Scorri ogni pixel dell'immagine con padding (escludendo i bordi aggiunti) e applica il filtro di media.

   *Per ogni pixel, calcola la somma dei prodotti tra i pixel della sottomatrice dim_filtroxdim_filtro centrata sul pixel corrente e il filtro.*
   
- Crea l'immagine filtrata: Crea un nuovo array NumPy che rappresenta l'immagine filtrata, dove ogni pixel contiene il risultato dell'applicazione del filtro.
- Visualizza l'immagine filtrata: Visualizza l'immagine filtrata utilizzando Matplotlib con la mappa di colori 'gray'.
- Aggiungi una colorbar per visualizzare la scala dei valori dell'immagine filtrata.
- Mostra l'immagine filtrata con Matplotlib.


Requisiti:

-  Utilizzare la libreria Matplotlib per caricare e visualizzare l'immagine.
-  Utilizzare la libreria NumPy per la manipolazione degli array e per il padding.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Carica l'immagine (assicurati che 'cameraman.png' sia nella stessa cartella dello script)
# plt.imread converte automaticamente l'immagine in un array NumPy 2D (se in scala di grigi)
image = plt.imread('cameraman.png')

# 2. Definisci la dimensione del filtro (deve essere un numero dispari, es. 3, 5, 7)
dim_filtro = 7
pad = dim_filtro // 2  # Calcola di quanti pixel dobbiamo allargare i bordi

# 3. Applica il padding con effetto specchio (quello che abbiamo visto nell'esercizio precedente!)
# Usiamo 'pad' per aggiungere il bordo su tutti i lati (sopra, sotto, destra, sinistra)
image_bordered = np.pad(image, pad_width=pad, mode='reflect')

# 4. Crea il filtro di media
# Una matrice dim_filtro x dim_filtro dove ogni cella vale 1 / (area_totale)
area_filtro = dim_filtro * dim_filtro
filtro_media = np.full((dim_filtro, dim_filtro), 1.0 / area_filtro)

# 5. Prepara un array vuoto (della stessa forma dell'immagine originale) per salvare il risultato
image_filtered = np.zeros_like(image)
altezza, larghezza = image.shape

# 6. Applica il filtro: scorriamo ogni pixel dell'immagine originale
for i in range(altezza):
    for j in range(larghezza):
        
        # Estraiamo la sottomatrice (la "finestra" che scorre) dall'immagine con il padding
        # Nota: gli indici i e j nell'immagine originale corrispondono al centro della 
        # sottomatrice nell'immagine con padding
        sottomatrice = image_bordered[i : i + dim_filtro, j : j + dim_filtro]
        
        # Calcoliamo la somma dei prodotti (prodotto element-wise e poi somma totale)
        nuovo_pixel = np.sum(sottomatrice * filtro_media)
        
        # Salviamo il nuovo valore
        image_filtered[i, j] = nuovo_pixel

# 7. Visualizzazione dei risultati
# Creiamo una figura con due sotto-grafici (subplot) per fare un confronto "Prima e Dopo"
plt.figure(figsize=(12, 6))

# Sotto-grafico 1: Immagine Originale
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Immagine Originale")
plt.axis('off') # Nascondiamo gli assi numerici per un effetto più pulito

# Sotto-grafico 2: Immagine Filtrata
plt.subplot(1, 2, 2)
# plt.imshow restituisce un oggetto che usiamo per generare la colorbar
im_plot = plt.imshow(image_filtered, cmap='gray')
plt.title(f"Immagine Filtrata (Filtro Media {dim_filtro}x{dim_filtro})")
plt.axis('off')

# 8. Aggiungi la colorbar riferita al grafico filtrato
plt.colorbar(im_plot, fraction=0.046, pad=0.04)

# Mostra il tutto a schermo
plt.tight_layout()
plt.show()
```
