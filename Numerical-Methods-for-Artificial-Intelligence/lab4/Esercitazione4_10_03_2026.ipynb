- **7.** Valutare in aritmetica floating point l'espressione
$$A(x)=\frac{1}{x}-\frac{1}{(x+1)}$$  nel vettore   $x=[10^0, 10^1,... ,10^{20}]$ contenente potenze successive  di $10$,  Si calcoli per ogni elemento di x l'errore relativo considerando come valore esatto quello calcolato usando  l'espressione algebricamente equivalente
$$
B(x)=\frac{1}{x (x+1)}.
$$ Si visualizzi l'errore relativo in un grafico in scala logaritmica su $x$ e $y$

```python
import numpy as np
import matplotlib.pyplot as plt

esponenti = np.arange(21, dtype=np.float64)
x = 10.0 ** esponenti

# 2. Valutazione delle espressioni
# Espressione A: calcolata come differenza di due frazioni
A = (1.0 / x) - (1.0 / (x + 1.0))

# Espressione B: calcolata in modo numericamente stabile (usata come valore esatto)
B = 1.0 / (x * (x + 1.0))

# 3. Calcolo dell'errore relativo
# Formula: |Approssimato - Esatto| / |Esatto|
errore_relativo = np.abs(A - B) / np.abs(B)

plt.figure(figsize=(10, 6))

plt.loglog(x, errore_relativo, marker='o', linestyle='-', color='red', label='Errore Relativo')

plt.title('Errore Relativo causato dalla Cancellazione Catastrofica')
plt.xlabel('x (Scala Logaritmica)')
plt.ylabel('Errore Relativo (Scala Logaritmica)')

plt.grid(True, which="both", ls="--", alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()

# Stampa di alcuni valori per verifica testuale
print("Analisi dei risultati per potenze elevate:")
for i in [10, 15, 16, 20]:
    print(f"x = 10^{i:2d} | Errore Relativo = {errore_relativo[i]:.2e}")
```

- **8.** Scrivere uno script Matlab che fornisca le approssimazioni di $e=\exp(1)$ ottenute con la formula
$f(n)=(1/n+1)^n$ per $n=10^k$ e  $k=0, ..., 16$. Visualizzare: un grafico con i valori di $f(n)$ al variare di n ed un grafico con i valori dell'errore relativo (considerando come valore esatto il valore math.exp(1)).  Cosa si osserva? Commentare i risultati ottenuti.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Vettore degli esponenti k (da 0 a 16) e calcolo di n
k = np.arange(0, 17, dtype=np.float64)
n = 10.0 ** k

# 2. Calcolo dell'approssimazione f(n)
f_n = (1.0 / n + 1.0) ** n

e_esatto = np.exp(1)

# 4. Calcolo dell'errore relativo
errore_relativo = np.abs(f_n - e_esatto) / e_esatto

# 5. Visualizzazione dei grafici
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Grafico 1: Andamento di f(n)
ax1.semilogx(n, f_n, marker='o', color='blue', label=r'$f(n) = (1 + 1/n)^n$')
ax1.axhline(e_esatto, color='green', linestyle='--', label=r'Valore esatto ($e$)')
ax1.set_title("Approssimazione di e al variare di n")
ax1.set_xlabel("n (Scala Logaritmica)")
ax1.set_ylabel("f(n)")
ax1.grid(True, which="both", ls="--", alpha=0.5)
ax1.legend()

# Grafico 2: Errore Relativo
ax2.loglog(n, errore_relativo, marker='s', color='red', label='Errore Relativo')
ax2.set_title("Errore Relativo dell'Approssimazione")
ax2.set_xlabel("n (Scala Logaritmica)")
ax2.set_ylabel("Errore Relativo (Scala Logaritmica)")
ax2.grid(True, which="both", ls="--", alpha=0.5)
ax2.legend()

plt.tight_layout()
plt.show()

# Stampa di alcuni valori chiave per l'analisi
print("Analisi dei risultati:")
for i in [1, 8, 15, 16]:
    print(f"k={i:2d} | n=10^{i:<2d} | f(n) = {f_n[i]:.8f} | Errore Rel. = {errore_relativo[i]:.2e}")
```
